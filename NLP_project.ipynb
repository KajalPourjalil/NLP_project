# Preprocessing and tokenization for squad dataset

```python
import pandas as pd
from transformers import BertTokenizerFast

# Load your data
train_df_1 = df1
val_df_1 = df2

# Initialize the tokenizer
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')

# Function to add start and end token positions for answers
def add_token_positions(df):
    contexts = df['context'].tolist()
    questions = df['question'].tolist()
    answers = df['answer'].tolist()
    answer_starts = df['answer_start'].tolist()

    encodings = tokenizer(
        questions,
        contexts,
        truncation=True,
        padding=True,
        return_offsets_mapping=True
    )

    start_positions = []
    end_positions = []

    for i in range(len(answers)):
        offsets = encodings['offset_mapping'][i]
        answer_start = answer_starts[i]
        answer_end = answer_start + len(answers[i])

        # Look for token indices corresponding to answer_start and answer_end
        start_idx = 0
        end_idx = 0

        for idx, (start, end) in enumerate(offsets):
            if start <= answer_start < end:
                start_idx = idx
            if start < answer_end <= end:
                end_idx = idx
                break

        start_positions.append(start_idx)
        end_positions.append(end_idx)

    # Remove offset mapping after extracting positions
    encodings.pop("offset_mapping")

    # Add start and end positions to encodings
    encodings.update({'start_positions': start_positions, 'end_positions': end_positions})

    return encodings

# Apply function to your datasets
train_encodings_1 = add_token_positions(train_df_1)
val_encodings_1 = add_token_positions(val_df_1)
```
